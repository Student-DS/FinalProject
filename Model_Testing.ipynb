```python
#defining function to fetch the data from the dataset according to the input given by the user
def filterdata(Test_od, Postcode, Rooms, EnergyRating,PropertyType,TotalArea):
    data = Test_od.query('Postcode == @Postcode & NUMBER_HABITABLE_ROOMS == @Rooms')
    return data
```

```python
def filterdata2(Test_ot, Postcode, Rooms, EnergyRating,PropertyType,TotalArea):
    data2 = Test_ot.query('Postcode == @Postcode & NUMBER_HABITABLE_ROOMS == @Rooms')
    return data2
```
